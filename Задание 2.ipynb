Контрольная работа состоит из пяти вопросов. На каждый вопрос приведено четыре варианта ответа, один из которых правильный. Случайная величина X задаёт число правильных ответов при простом угадывании. Найдите математическое ожидание данной случайной величины. Найдите дисперсию случайной величины X.

Решение: <br>
всего сочетаний из 4х ответов: $C_4^1=\frac{4!}{1!(4-1)!} = \frac{24}{6} = 4$<br>
сочетаний из одного ответа: $C_1^1=\frac{1!}{!1(1-1)!} = 1$<br>
вероятность выбора правильного ответа $P(X)=\frac{C_1^1}{C_4^1} = \frac{1}{4}$<br>
возможные значения случайной величины (количество правильных ответов при на 5 вопросов):<br>
$x_1 = 0$ ни одного правильного<br> 
$x_2 = 1$<br>
$x_3 = 2$<br>
$x_4 = 3$<br>
$x_5 = 4$ <br>
$x_6 = 5$ все правильные<br>
подсчитаем распределение при ответе на 5 вопросов для каждого значения случайной величины: $P(X = 0) = C_n^k p^k (1 - p)^{n - k}$, где $n=5, k=x_i, p=\frac{1}{4}$:<br><br>
$P(X = 0) = C_5^0*(\frac{1}{4})^0*(1-\frac{1}{4})^{5-0}=\frac{5!}{0!(5-0)!}*1*(\frac{3}{4})^5=1*1*\frac{243}{1024}=0.2373$<br><br>
$P(X=1)=C_5^1*(\frac{1}{4})^1*(1-\frac{1}{4})^{5-1} = \frac{5!}{1!(5-1)!}*\frac{1}{4}*(\frac{3}{4})^4=5*\frac{1}{4}*\frac{81}{256}=0.3955$<br><br>
$P(X=2)=C_5^2*(\frac{1}{4})^2*(1-\frac{1}{4})^{5-2} = \frac{5!}{2!(5-2)!}*\frac{1}{16}*(\frac{3}{4})^3=10*\frac{1}{16}*\frac{27}{64}=0.2636$<br><br>
$P(X=3)=C_5^3*(\frac{1}{4})^3*(1-\frac{1}{4})^{5-3} = \frac{5!}{3!(5-3)!}*\frac{1}{64}*(\frac{3}{4})^2=10*\frac{1}{64}*\frac{9}{16}=0.0877$<br><br>
$P(X=4)=C_5^4*(\frac{1}{4})^4*(1-\frac{1}{4})^{5-4} = \frac{5!}{4!(5-4)!}*\frac{1}{256}*(\frac{3}{4})^1=5*\frac{1}{256}*\frac{3}{4}=0.0146$<br><br>
$P(X=5)=C_5^5*(\frac{1}{4})^5*(1-\frac{1}{4})^{5-5} =\frac{5!}{5!(5-5)!}*\frac{1}{1024}*(\frac{3}{4})^0=1*\frac{1}{1024}*1=0.0009$<br><br>
проверим: 0.2373 + 0.3955 + 0.2636 + 0.0877 + 0.0146 + 0.0009 = 0,999<br><br>
$M(X)=np$, где $p=\frac{1}{4}, n = 5, M(X) = 5 * 0.25 = 1.25$<br>
$D(X) = np(1-p) = 5*0.25(1-0.25)=1.25*0.75=0.9375$<br><br>
Ответ: $M(X)=1.25, D(X)=0,9375$

#### Задача 4
Производятся выстрелы по мишени. Вероятность попадания в мишень при одном выстреле равна 0.01. Сколько выстрелов нужно сделать чтобы быть уверенным с вероятностью 0.9, что хотя бы 6 раз будет совершено попадание?

Решение:<br>
$P(X=6) = 0.9$ - искомая вероятность попадания 6 выстрелов в мишень при n выстрелах<br>
$P(X = 1)=0.01$ - вероятность попадания в мишень одного выстрела<br>
Примем за событие А 100 выстрелов. Тогда вероятность попадания в мишень $P(X=100)=p*100 =1$, то есть 1 попадание в мишень это 100 выстрелов. По условию нам нужно вычислить количество n выстрелов для 6 попаданий. Следовательно, необходимо вычислить распределение для 6 событий А. При этом, поскольку вероятность каждого события А равна 1, то возможные сочетания 6 событий А $x_1=6$ (6 по 100 выстрелов с вероятностью попадания 1 каждые 100 выстрелов).<br>
По формуле биномиального распределения $P(X = k) = C_n^k p^k (1 - p)^{n - k}$, где к = 6, p = 1, n = 6:<br><br>
$P(X=6)=C_6^6*p^1*(1-1)^{6-6}=\frac{6!}{6!(6-6)!}*1^1(1-1)^0=1$<br><br>
С вероятностью 1 после 6*100=600 выстрелов произойдёт попадание по мишени. Вычислим вероятность 0,9. Вероятность 1 так относится к 600 выстрелам как вероятность 0,9 к n выстрелам:<br><br>
$\frac{1}{600}=\frac{0.9}{n} => n = \frac{600 * 0.9}{1} = 540$<br><br>
Ответ: вероятность 0,9 попадания 6 выстрелов в мишень достигается при 540 выстрелах.
