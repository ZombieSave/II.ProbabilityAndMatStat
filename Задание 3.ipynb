#### Задача 1


Даны значения зарплат из выборки выпускников:

100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 230, 24, 57, 55, 70, 75, 65, 84, 90, 150

Используя только встроенные питоновские функции и структуры данных (т.е. без библиотек numpy, pandas и др.), посчитайте (несмещённое) среднее квадратичное отклонение для данной выборки.

```python
import math

wages = [100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 230, 24, 57, 55, 70, 75, 65, 84, 90, 150]
n = len(wages) # количество элементов выборки
k = 1/(n-1) # коэффициент перед суммой
X = sum(wages)/n # выборочное среднее 
sigma = math.sqrt(sum([(x-X)**2 for x in wages])*k) # среднее квадратичное отклонение

print(f"Ответ: несмещённое среднее квадратичное отклонение равно {sigma}")
```

#### Задача 2

Найдите число выбросов в выборке из задачи 1. Для определения выбросов используйте методику как при построении “усов” в boxplot, однако, как и в задаче 1, пользоваться можно только встроенными функциями и структурами данных.


```python
wages = [100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 230, 24, 57, 55, 70, 75, 65, 84, 90, 150]
wages.sort()
print("Отсортированный по возрастанию список зарплат:")
print(wages)
length = len(wages)

# медиана - это 2q квартиль. найдём для нечётной выборки
median_index = length // 2
Q2 = wages[median_index]

# найдём 1й квартиль - посередине слева от медианы
i = (length // 4)
Q1 = wages[i]

#найдём третий квартить
i = (length // 4)*3
Q3 = wages[i]

IQR = Q3 - Q1

print(f"первый квартиль: {Q1}\nвторой квартиль: {Q2}\nтретий квартиль:{Q3}\nинтерквартильный размах: {IQR}")
print("Сверял с pandas: df.quantile(0.25), df.quantile(0.50), df.quantile(0.75)")

# левый "ус"
L = Q1 - 1.5*IQR
# правый "ус"
R = Q3 + 1.5*IQR

L_len = len(list(filter(lambda x: x < L, wages)))
R_len = len(list(filter(lambda x: x > R, wages)))
print(f"\nОтвет: количество выбросов: {L_len + R_len}")
```

#### Задача 3

В университет на факультет A поступило 100 человек, на факультет B - 90 человек и на факультет C - 45 человек. 
Вероятность того, что студент с факультета A сдаст первую сессию, равна 0.6. 
Для студента с факультета B эта вероятность равна 0.7, а для студента с факультета C - 0.8. 
Случайным образом выбирается студент, сдавший первую сессию. Какое событие наиболее вероятно:<br>
студент учится на факультете A,<br>
студент учится на факультете B,<br>
студент учится на факультете C?


Решение: применим формулу Байеса $P(A|S) = \frac{P(S|A)*P(A)}{P(S)}$ для каждой группы. При этом общее количество студентов на трёх факультетах: N = 100 + 90 + 45 = 235<br><br>

Считаем для группы A:<br>
$N_A=100$ - количество студентов в группе<br><br>
$P(A)_A = ((100*100)/235)/100=0.42$ - вероятность того что студетн из первой группы<br><br>
$P(S|A)_A = 0.6$<br><br>


Считаем для группы B:<br>
$N_B=90$ - количество студентов в группе<br><br>
$P(A)_B=((90*100)/235)/100=0.38$ - вероятность того что студетн из второй группы<br><br>
$P(S|A)_B = 0.7$<br><br>



Считаем для группы C:<br>
$N_C=45$ - количество студентов в группе<br><br>
$P(A)_C=((45*100)/235)/100 = 0.19$ - вероятность того что студетн из третьей группы<br><br>
$P(S|A)_C = 0.8$<br><br>



Полная вероятность:
$P(S) = P(S|A)_A*P(A)_A + P(S|A)_B*P(A)_B + P(S|A)_C*P(A)_C = 0.6*0.42 + 0.7*0.38 + 0.8*0.19=0.67$<br>

$P(A|S)_A=\frac{0.6*0.42}{0.67}=0.376$<br><br>
$P(A|S)_B=\frac{0.7*0.38}{0.67}=0.397$<br><br>
$P(A|S)_C = \frac{0.8*0.19}{0.67}=0.226$<br><br>

Ответ: наиболее вероятно (0.397) что студент, сдавший сессию, учится на факультете B.
